```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import c

# Problem 1
def ctprime_axis(x, v):
    beta = v / c
    return x / beta
    
# Problem 2
def xprime_axis(x, v):
    beta = v / c
    return beta * x



#Problem 3
v1 = 0.2 * c
ct_for_ctprime = ctprime_axis(x_vals, v1)
ct_for_xprime  = xprime_axis(x_vals, v1)

plt.figure(figsize=(7,7))
plt.plot(x_vals, ct_for_ctprime, 'r', label=r"$ct' \; (v=0.2c)$")
plt.plot(x_vals, ct_for_xprime, 'r--', label=r"$x' \; (v=0.2c)$")

ax = plt.gca()
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_aspect('equal')
plt.xlim([-1e10, 1e10])
plt.ylim([-1e10, 1e10])
plt.xlabel(r"$x \,(\mathrm{m})$", color='black')
plt.ylabel(r"$ct \,(\mathrm{m})$", color='black')
plt.legend(loc='upper left')
plt.title("Problem 3")
plt.show()

# Problem 4
velocities_q4 = [0.25*c, 0.5*c, 0.75*c]
colors_q4 = ['blue', 'green', 'orange']

plt.figure(figsize=(7,7))
plt.plot(x_vals, np.zeros_like(x_vals), 'k', label=r"$x$ axis")
plt.plot(np.zeros_like(x_vals), x_vals, 'k', label=r"$ct$ axis")

for v, col in zip(velocities_q4, colors_q4):
    ct_line = ctprime_axis(x_vals, v)
    x_line = xprime_axis(x_vals, v)
    plt.plot(x_vals, ct_line, color=col, label=rf"$ct' \;(v={v/c:.2f}c)$")
    plt.plot(x_vals, x_line, color=col, linestyle='--', label=rf"$x' \;(v={v/c:.2f}c)$")

ax = plt.gca()
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_aspect('equal')
plt.xlim([-1e10, 1e10])
plt.ylim([-1e10, 1e10])
plt.xlabel(r"$x \,(\mathrm{m})$")
plt.ylabel(r"$ct \,(\mathrm{m})$")
plt.legend(loc='upper left', fontsize=10)
plt.title("Problem 4")
plt.show()

# Problem 5
velocities_q5 = [0.5*c, 0.75*c, 0.875*c, 0.9375*c]
gray_shades = ['0.7', '0.5', '0.3', '0.1']

plt.figure(figsize=(7,7))
plt.plot(x_vals, np.zeros_like(x_vals), 'k', label=r"$x$ axis")
plt.plot(np.zeros_like(x_vals), x_vals, 'k', label=r"$ct$ axis")

for v, shade in zip(velocities_q5, gray_shades):
    ct_line = ctprime_axis(x_vals, v)
    x_line = xprime_axis(x_vals, v)
    plt.plot(x_vals, ct_line, color=shade, label=rf"$ct' \;(v={v/c:.4f}c)$")
    plt.plot(x_vals, x_line, color=shade, linestyle='--', label=rf"$x' \;(v={v/c:.4f}c)$")

plt.plot(x_vals, x_vals, 'k:', label=r"Limiting world line ($v=c$)")

ax = plt.gca()
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_aspect('equal')
plt.xlim([-1e10, 1e10])
plt.ylim([-1e10, 1e10])
plt.xlabel(r"$x \,(\mathrm{m})$")
plt.ylabel(r"$ct \,(\mathrm{m})$")
plt.legend(loc='upper left', fontsize=9)
plt.title("Problem 5")
plt.show()
```
