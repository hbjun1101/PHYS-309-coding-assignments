```python
#Question 0
import numpy as np
import matplotlib.pyplot as plt
from qmsolve import Hamiltonian, SingleParticle, init_visualization, , eV
def harmonic_oscillator(particle):
    k = 100 * eV / **2
    return 0.5 * k * particle.x**2
H = Hamiltonian(
    particles=SingleParticle(),
    potential=harmonic_oscillator,
    spatial_ndim=1, N=512, extent=20*
)
eigenstates = H.solve(max_states=30)
print(eigenstates.energies)
vis = init_visualization(eigenstates)
vis.slider_plot()
# assuming `eigenstates` and `vis` from above
for n in range(5):
    vis.plot_eigenstate(n)
plt.show()
```
