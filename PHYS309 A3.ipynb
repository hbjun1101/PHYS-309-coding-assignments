```python
#Question 1
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

c = constants.c
u_prime = 0.99 * c
theta = np.radians(15)

v = np.linspace(0, 0.99 * c, 100)

# 1a:
u = (u_prime + v) / (1 + (v * u_prime) / c**2)

plt.figure(figsize=(10, 6))
plt.plot(v / c, u / c, label="Proton speed")
plt.xlabel("Spacecraft speed / c", fontsize=14)
plt.ylabel("Proton speed / c", fontsize=14)
plt.title("Q1-a Proton Speed Measured by Frank", fontsize=16)
plt.grid(True)
plt.legend()
plt.show()

# 1b:
u_x_prime = u_prime * np.cos(theta)
u_y_prime = u_prime * np.sin(theta)
gamma = 1 / np.sqrt(1 - (v**2 / c**2))
u_y = u_y_prime / (gamma * (1 + v * u_x_prime / c**2))

plt.figure(figsize=(10, 6))
plt.plot(v / c, u_y / c, label="Proton y-velocity", color='orange')
plt.xlabel("Spacecraft speed / c", fontsize=14)
plt.ylabel("Proton y-velocity / c", fontsize=14)
plt.title("Q1-b Y-Component of Proton Velocity Measured by Frank", fontsize=16)
plt.grid(True)
plt.legend()
plt.show()

#Question 2
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

c = constants.c  # m/s
t = np.linspace(0, 2e-3, 100)
x_flash1 = c * t
x_flash2 = 300000e3 - c * t
x_probe = 500000e3 - 0.1 * c * t

plt.figure(figsize=(10, 6))
plt.plot([0], [0], 'ro', label="Flash at x = 0 km")
plt.plot([300000e3], [0], 'bo', label="Flash at x = 300000 km")
plt.plot(x_flash1, c * t, 'r--', label="Light from flash 1")
plt.plot(x_flash2, c * t, 'b--', label="Light from flash 2")
plt.plot([150000e3], [0], 'g*', label="JWST")
plt.plot([150000e3] * len(t), c * t, 'g-', label="JWST world line")
plt.plot(x_probe, c * t, 'm-', label="Probe world line")
plt.plot([500000e3], [0], 'mo', label="Probe at t = 0")

plt.xlabel("x (km)", fontsize=14)
plt.ylabel("ct (km)", fontsize=14)
plt.title("Q2 Spacetime Diagram", fontsize=16)
plt.legend()
plt.grid(True)
plt.show()

#Question 3
import numpy as np
import matplotlib.pyplot as plt

b = 2.898e-3
stars = ["O5", "B5", "A5", "F5", "G5", "K5", "M5"]
temps = [40000, 15200, 8310, 6700, 5660, 4400, 3200]
lambda_max = b / np.array(temps) * 1e9

plt.figure(figsize=(10, 6))
plt.scatter(temps, lambda_max, color='blue')
for i, star in enumerate(stars):
    plt.text(temps[i], lambda_max[i], star, fontsize=12)
plt.xlabel("Temperature (K)", fontsize=14)
plt.ylabel("Wavelength of max intensity (nm)", fontsize=14)
plt.title("Q3 Wien's Displacement Law for Stars", fontsize=16)
plt.grid(True)
plt.show()

#Question 4
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

h = constants.h
e = constants.e
nu = np.linspace(0, 2e15, 100)
phi_Al = 4.08
phi_Au = 5.1
K_Al = (h * nu / e) - phi_Al
K_Au = (h * nu / e) - phi_Au

plt.figure(figsize=(10, 6))
plt.plot(nu/1e15, K_Al, label="Aluminum ( = 4.08 eV)")
plt.plot(nu/1e15, K_Au, label="Gold ( = 5.1 eV)")
plt.axhline(0, color='black', linestyle='--')
plt.xlabel("Frequency (10 Hz)", fontsize=14)
plt.ylabel("Maximum Kinetic Energy (eV)", fontsize=14)
plt.title("Q4 Photoelectric Effect", fontsize=16)
plt.legend()
plt.grid(True)
plt.ylim(-1, 10)
plt.show()
```
