```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants
from mpl_toolkits.axes_grid1 import make_axes_locatable

E_R = constants.physical_constants['Rydberg constant times hc in eV'][0]

# Question 1
def photon_energy(Z, n):
    return Z**2 * E_R * (1/n**2 - 1/(n+1)**2)

# Question 2
n = np.arange(1, 11)
Z = np.array([1, 2, 3])

# Question 3
energies = photon_energy(Z[:, None], n[None, :])

# Question 4
Nv, Zv = np.meshgrid(n, Z)
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
plot1 = ax.scatter(Nv.flatten(), Zv.flatten(), c=energies.flatten(), cmap='viridis', s=100)
ax.set_aspect('equal')

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="8%", pad=0.1)
cbar = plt.colorbar(plot1, cax=cax)
cbar.ax.tick_params(labelsize=12)
cbar.set_label('(eV)', fontsize=12)

ax.set_xlabel('Energy Level n', fontsize=14)
ax.set_ylabel('Z', fontsize=14)
ax.tick_params(labelsize=12)
ax.set_xticks(np.arange(1, 11))
ax.set_yticks([1, 2, 3])

plt.show()
```
